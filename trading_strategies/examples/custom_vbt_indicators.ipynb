# Overview

This notebook shows how to use custom vectorbt indicators.

```python
import pandas as pd
import datetime as dt

# Define a time series.
signal = pd.DataFrame({
    'p': [1, 2, 3, 4, 5],
}, index=pd.Index([
    dt.datetime(2020, 1, 1),
    dt.datetime(2020, 1, 2),
    dt.datetime(2020, 1, 3),
    dt.datetime(2020, 1, 4),
    dt.datetime(2020, 1, 5),
])).astype(float)

signal
```

```python
import ge.trading.indicators as indicators

# Run the custom defined MedianIndicator over the signal with two different values for the 'window'
# parameter. This yields a DataFrame-like object.
v = indicators.MedianIndicator.run(signal, [3, 4])

v.median
```

```python
# Having been defined using the IndicatorFactory, the MedianIndicator inherits several built in
# features that did not have to be defined explicitly. The following method median_crossed_above
# is automatically generated by the IndicatorFactory.
v.median_crossed_above(2.5)
```
